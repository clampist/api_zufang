```python
def parse_response(resp):
    next_page = resp.get('Page').get('next_page')
    room_list = []
    for room_dict in resp.get('Post'):
        room_pic = room_dict.get('room_pic')
        if room_pic:
            one_room = room_dict.get('one_room')
            if len(one_room) > 0 and one_room[0] == '1':
                one_or_not = '1'
            else:
                one_or_not = '0'
            file_ext = room_pic.split('.')[-1]
            room_id = room_dict.get('url').split('/')[-1].split('.')[0]
            pic_file_name = one_or_not + '.' + room_id + '.' + file_ext
            res_dict = (pic_file_name, room_pic)
            room_list.append(res_dict)
    return next_page, room_list
```

```python
def download_pic(pic_list):
    for file_name, pic_url in pic_list:
        big_pic_url = pic_url.replace('200x150', '600x450')
        big_img = requests.get(big_pic_url)
        if big_img.status_code == requests.codes.ok:
            img = big_img
            file_name = 'big.' + file_name
        else:
            file_name = 'small.' + file_name
            img = requests.get(pic_url)
            if img.status_code != requests.codes.ok:
                continue
        if not os.path.exists(file_name):
            try:
                with open('output/' + file_name, 'wb') as localFile:
                    localFile.write(img.content)
            except Exception as e:
                print(e)
                print(file_name)
```
